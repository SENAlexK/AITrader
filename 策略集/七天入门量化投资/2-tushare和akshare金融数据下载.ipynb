```python

# 导入tushare
import tushare as ts
# 初始化pro接口
pro = ts.pro_api('your token here')
```

```python
def download_data(symbol):
    # 拉取数据
    df = pro.index_daily(**{
        "ts_code": symbol,
        "trade_date": "",
        "start_date": "",
        "end_date": "",
        "limit": "",
        "offset": ""
    }, fields=[
        "ts_code",
        "trade_date",
        "close",
        "open",
        "high",
        "low",
        "pre_close",
        "change",
        "pct_chg",
        "vol",
        "amount"
    ])
    return df

symbol = '000300.SH'
df = download_data(symbol)

# 修改列名
df.rename(columns={'ts_code':'symbol','trade_date':'date','vol':'volume'},inplace=True)
df
```

```python
# 保存到本地csv文件
df.to_csv(f'data/{symbol}.csv',index=False)
```
